```python
import numpy as np
import torch
from torch.nn.functional import softmax
import matplotlib.pyplot as plt

from data import SampleClips
```

```python
N = 200
frame_nums = np.arange(start=0, stop=N-1, dtype=float)
```

```python
probs = frame_nums/frame_nums.sum() 
probs_c = -probs + probs[-1]
probs_c.sum()
```

```python
counts = {
    int(f) + 1: 0 for f in frame_nums
}
reps = 30
for x in range(reps):
    f_rand = np.random.choice(frame_nums, N-1, p = probs_c)
    f_rand_1 = f_rand + 1
    for item in f_rand_1:
        counts[int(item)] += 1

    
exp_ps = np.array(list(counts.values()))/(reps * (N-1))
plt.plot(exp_ps)
plt.plot(probs_c, color="black")
```
